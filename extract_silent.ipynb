The cell below memory maps the file in order `'C'` and then loads the new memory mapped file. The saved files from motion correction are memory mapped files stored in `'F'` order. Their paths are stored in `mc.mmap_file`.

```python
#%% MEMORY MAPPING
border_to_0 = 0 if mc.border_nan is 'copy' else mc.border_to_0 #what to do with borders from motion correction 
fname_new = cm.save_memmap(mc.mmap_file, base_name=TSeries_2+'_memmap_', order='C',# memory map the file in order 'C'
                           border_to_0=border_to_0, dview=dview) # exclude borders

# now load the file
Yr, dims, T = cm.load_memmap(fname_new)
images = np.reshape(Yr.T, [T] + list(dims), order='F') #load frames in python format (T x X x Y)
```

```python
#%% restart cluster to clean up memory
cm.stop_server(dview=dview)
c, dview, n_processes = cm.cluster.setup_cluster(
    backend='local', n_processes=None, single_thread=False)
```

### run seeded CNMF for to get rois + timeseries  for cells silent in session 2

load cnmf params Amy used

```python
# dataset dependent parameters

rf = None                   # half-size of the patches in pixels. Should be `None` when seeded CNMF is used.
only_init = False           # has to be `False` when seeded CNMF is used
gSig = cnm1.params.init['gSig']            # expected half size of neurons in pixels, very important for proper component detection

# params object
opts_dict = {'fnames': fname_new,
            'decay_time': cnm1.params.data['decay_time'],
            'p': cnm1.params.temporal['p'],
            'nb': cnm1.params.temporal['nb'],
            'rf': rf,
            'only_init': only_init,
            'gSig': gSig,
            'ssub': cnm1.params.init['ssub'],
            'tsub': cnm1.params.init['tsub'],
            'merge_thr': cnm1.params.merging['merge_thr']}

opts.change_params(opts_dict);
```

run seeded cnmf 

```python
cnm_seeded = cnmf.CNMF(n_processes, params=opts, dview=dview, Ain=A1_silent)
cnm_seeded.fit(images)
```

plot components

```python
CI = cm.local_correlations(images.transpose(1,2,0))     #  correlation image
CI[np.isnan(CI)] = 0
cnm_seeded.estimates.plot_contours_nb(img=CI)
```
